```python
#deletes if the start or end station is null 
df23 = df23[(~df23['start_station_id'].isnull()) & (~df23['end_station_id'].isnull())]
```

```python
# Convert 'started_at' and 'ended_at' to datetime if they are not already
df23['started_at'] = pd.to_datetime(df23['started_at'])
df23['ended_at'] = pd.to_datetime(df23['ended_at'])

# Calculate the duration in seconds
df23['duration'] = (df23['ended_at'] - df23['started_at']).dt.total_seconds()

# Now df23 contains a new field 'duration' representing the duration between 'started_at' and 'ended_at'
#print(df23[['started_at', 'ended_at', 'duration']])
```

```python
station_df = df23[['start_station_name', 'start_station_id', 'start_lat', 'start_lng']].copy()
station_df.rename(columns={'start_station_name': 'station_name', 
                       'start_station_id': 'station_id', 
                       'start_lat': 'station_lat', 
                       'start_lng': 'station_lng'}, inplace=True)

# Make the DataFrame distinct based on station_id
station_df = station_df.drop_duplicates(subset=['station_id'])

# Now new_df contains distinct rows based on station_id
print(station_df)
```

```python
# Load the GeoJSON file containing the boundaries of New York City boroughs from https://data.cityofnewyork.us/City-Government/Borough-Boundaries/tqmj-j8zm
nyc_boroughs = gpd.read_file(r'C:\Users\david\OneDrive\Documents\GitHub\Tableau_citi_bike\2023-citibike-tripdata\2023-citibike-tripdata\BoroughBoundaries.geojson')

# Define a function to determine the borough based on latitude and longitude
def get_borough(lat, lon):
    point = Point(lon, lat)
    for index, row in nyc_boroughs.iterrows():
        if point.within(row['geometry']):
            return row['boro_name']
    return 'Unknown'

# Apply the get_borough function to each row to get start and end boroughs
station_df['borough'] = station_df.apply(lambda row: get_borough(row['station_lat'], row['station_lng']), axis=1)

```
